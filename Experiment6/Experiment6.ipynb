```python
import math
import cv2
import numpy as np
```

```python
img = cv2.imread('input/ricegrains_mono.bmp')

# convert returned numpy array to list
img = img[:,:,0]
img = [[complex(img[i][j]) for j in range(len(img[0]))] for i in range(len(img))]
```

```python
# Kernels 

kernel1=[[1,1]]

kernel2=[[1,1,1],
          [1,1,1],
          [1,1,1]]

kernel3=[[0,1,0],
          [1,1,1],
          [0,1,0]]

def kernel_gen(dim):
  kernel=[[1 for _ in range(dim)] for __ in range(dim)]
  return kernel

def withinRange(i,j,n,m):
  return i>=0 and i<n and j>=0  and j<m
```

```python
# Implements the Dilation Operation

def DilateBinary(x,b):
  y=[[0 for _ in range(len(x))] for __ in range(len(x[0]))]

  for i in range(len(x)):
    for j in range(len(x[0])):

      if x[i][j]==0:
        continue

      for i1 in range(len(b)):
        for j1 in range(len(b[0])):

          if(b[i1][j1]==0):
            continue

          x_coord=i+i1-int(len(b)/2)
          y_coord=j+j1-int(len(b[0])/2)

          if withinRange(x_coord,y_coord,len(x),len(x[0])):
            y[x_coord][y_coord]=255          
    
  return y
```

```python
# Implements the Erosion Operation

def ErodeBinary(x,b):
  y=[[0 for _ in range(len(x))] for __ in range(len(x[0]))]

  for i in range(len(x)):
    for j in range(len(x[0])):
      allWhite=True
      for i1 in range(len(b)):
        for j1 in range(len(b[0])):
          
          if(b[i1][j1]==0):
            continue
          x_coord=i+i1-int(len(b)/2)
          y_coord=j+j1-int(len(b[0])/2)

          if withinRange(x_coord,y_coord,len(x),len(x[0])) and x[x_coord][y_coord]==255:
            continue
          else:
            allWhite=False
        
      if allWhite:
        y[i][j]=255
  
  return y
```

```python
# Main Menu

while(True):
  print("Press 1 for kernel_1")
  print("Press 2 for kernel_2")
  print("Press 3 for kernel_3")
  print("Press 4 for custom_kernel")
  print("Press 5 to exit")
  op1=int(input("Choose Option: "))
  kernel=list()
  
  filename=""
  if op1==1:
    kernel=kernel1
    filename="kernel1"
  elif op1==2:
    kernel=kernel2
    filename="kernel2"
  elif op1==3:
    kernel=kernel3
    filename="kernel3"
  elif op1==4:
    dim=int(input("Enter dimension of kernel"))
    kernel=kernel_gen(dim)
    filename=f'kernel_{dim}'
  elif op1==5:
    break
  else:
    continue
  
  print(" Choose Operation ")
  print("\n")
  print("Press 1 for Erode Image")
  print("Press 2 for Dilate Image")
  print("Press 3 for Opening Image")
  print("Press 4 for Closing Image")

  op1=int(input("Choose Option: "))
  
  operatedImage=list()
  if op1==1:
    filename="output/Erosion/Erode_"+filename
    operatedImage=ErodeBinary(img,kernel)
  elif op1==2:
    filename="output/Dilation/Dilate_"+filename
    operatedImage=DilateBinary(img,kernel)
  elif op1==3:
    filename="output/Opening/Opening_"+filename
    operatedImage=ErodeBinary(img,kernel)
    operatedImage=DilateBinary(operatedImage,kernel)
  elif op1==4:
    filename="output/Closing/Closing_"+filename
    operatedImage=DilateBinary(img,kernel)
    operatedImage=ErodeBinary(operatedImage,kernel)
  else:
    print("Invalid Operation")
    continue
  
  filename=filename+".jpg"

  cv2.imwrite(filename,np.array(operatedImage).astype(np.uint8))
  print("\n\n")
```
